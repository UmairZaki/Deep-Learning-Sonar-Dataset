```python
import keras
import numpy as np
import pandas as pd
from keras.layers import Input, Dense
from keras.models import Model
from keras import layers, optimizers, losses, metrics

# load dataset
dataframe = pd.read_csv("sonar.csv", header=None)
dataset = dataframe.values

# shuffle data to shuffle rocks & mine data
np.random.shuffle(dataset)

# split into input (X) and output (Y) variables
x = dataset[:,0:60].astype(float)
y = dataset[:,60]

#normaliazing data
x -= x.mean(axis=0)
x /= x.std(axis=0)

#encode label in 1 and 0
a = y.tolist()
b= []
for i in a:
    if i == 'R':
        i = 1
        b.append(i)
    else:
        i = 0
        b.append(i)
y = np.asarray(b)

#split data in test and train
x_test = x[:45]
x_train = x[45:]
y_test = y[:45]
y_train = y[45:]

# split validation data
x_val = x_train[:45]
partial_x_train = x_train[45:]
y_val = y_train[:45]
partial_y_train = y_train[45:]

#model subclassing
class myModel(keras.Model):
    
    def __init__ (self):
        super(myModel, self).__init__()
        self.dense1 = Dense(60, activation='relu')
        self.dense2 = Dense(1, activation = 'sigmoid')
        
    def call(self, x):
        x = self.dense1(x)
        return self.dense2(x)
    
model = myModel()
sgd = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
loss = losses.binary_crossentropy
metrics = [metrics.binary_accuracy]

#compiler
model.compile(sgd,loss,metrics)

#train model
history = model.fit(partial_x_train,partial_y_train,epochs=100,batch_size=5,validation_data=(x_val, y_val))
```

```python
#evaluate on test data
results = model.evaluate(x_test, y_test)
print(results)
```
