```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers

# load dataset
dataframe = pd.read_csv("sonar.csv", header=None)
dataset = dataframe.values 

# shuffle data for shaffling rocks and mine
np.random.shuffle(dataset)

# split into input (X) and output (Y) variables
x = dataset[:,0:60].astype(float)
y = dataset[:,60]

#normaliazing data
x -= x.mean(axis=0)
x /= x.std(axis=0)

#encode label in 1 and 0
a = y.tolist()
b= []
for i in a:
    if i == 'R':
        i = 1
        b.append(i)
    else:
        i = 0
        b.append(i)
y = np.asarray(b)

# split data test and train
x_test = x[:45]
x_train = x[45:]
y_test = y[:45]
y_train = y[45:]

# split train data in partial tran and validation
x_val = x_train[:45]
partial_x_train = x_train[45:]
y_val = y_train[:45]
partial_y_train = y_train[45:]

#Keras Sequential API
model = Sequential()
model.add(Dense(60, activation='relu', input_shape=(60,)))
model.add(Dense(1, activation='sigmoid'))
sgd = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(optimizer=sgd,loss='binary_crossentropy',metrics=['accuracy'])
history = model.fit(partial_x_train,partial_y_train,epochs=100,batch_size=5,validation_data=(x_val, y_val))
```

```python
#evaluate on test data
results = model.evaluate(x_test, y_test)
print(results)
```
